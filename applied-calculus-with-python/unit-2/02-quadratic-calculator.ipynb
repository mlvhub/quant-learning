```python
from numpy import sqrt
from sympy.plotting import *
from sympy import Symbol
```

```python
# ax^2 + bx + c
def zeros(a, b, c):
    D = sqrt(b*b-4*a*c)
    x1 = (-b + D) / (2 * a)
    x2 = (-b - D) / (2 * a)

    return (x1, x2)
```

```python
def print_graph(a, b, c):
    x = Symbol("x")
    plot(a*x**2 + b*x + c)
```

```python
a = 1
b = 2
c = 1

zeros(a, b, c)
```

```python
print_graph(a, b, c)
```

```python
def linear(x1, y1, x2, y2):
    x = Symbol('x')
    m = (y2-y1)/(x2-x1)
    return m*(x - x1) + y1
```

```python
linear(1, -4, 3, 6)
```

```python
linear(4, 9, 1.5, 7)
```

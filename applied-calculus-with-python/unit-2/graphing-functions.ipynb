```python
from sympy.plotting import *
from sympy import Symbol
```

```python
x = Symbol("x")
```

```python
plot(x**2)
```

```python
plot1 = plot(x**2, (x, -3, 3))
plot1[0].line_color ='r'
```

```python
f1 = 3*x+2
f2 = x**3
plot(f1, f2)
```

```python
# piecewise function
plot((f1, (x, -10, 0)), (f2, (x, 0, 5)))
```
